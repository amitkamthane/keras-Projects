***Description about  AVILA Dataset***

- The Avila dataset has been extracted from 800 images of the Avila Bible,
- The dataset consists of various features about the images of the text, such as intercolumnar distance and margins of the text.
- The dataset also contains a class label that indicates if the pattern of the image falls into the most frequently occurring category or not.
- The palaeographic analysis of the  manuscript has  individuated the presence of 12 copyists. 
- The pages written by each copyist are not equally numerous.
- Each pattern contains 10 features and corresponds to a group of 4 consecutive rows.- 
- The prediction task consists in associating each pattern to one of the 12 copyists (labeled as: A, B, C, D, E, F, G, H, I, W, X, Y).

#### Define the keras model

- Define a function that returns a Keras model with three hidden layers.
- The first of size 10,  the second of size 6, and the third of size 4, 
- Apply L2 weight regularization and a ReLU activation function on each hidden layer. 
- Compile the model with the given parameters and return it from the model:


```python
import keras
```

```python
from keras.layers import Dense
from keras.models import Sequential
from keras.optimizers import SGD
from keras.regularizers import l2
from keras.losses import binary_crossentropy
```

```python
def build_model(lambda_parameter):
    model = Sequential()
    model.add(Dense(10,input_dim=X.shape[1],activation="relu",kernel_regularizer = l2(lambda_parameter)))
    model.add(Dense(6,activation="relu",kernel_regularizer=l2(lambda_parameter)))
    model.add(Dense(4,activation="relu",kernel_regularizer=l2(lambda_parameter)))
    model.add(Dense(1,activation="sigmoid"))
    model.compile(loss='binary_crossentropy',optimizer="sgd",metrics="accuracy")
    return model
```

#### Define a seed

- define a seed for random number generator so the result will be reproducible

#### Cereate Keras wrapper with scikit learn 

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = KerasClassifier(build_fn=build_model,verbose=1, shuffle=False)
```

#### Define differnt values of hyperparamter

```python
lambda_parameter = [0.01,0.5, 1]
```

```python
epochs = [50,100]
```

```python
batch_size=[20]
```

#### Create dictionary of hypermeter

```python
param_grid = dict(lambda_parameter= lambda_parameter,epochs=epochs,batch_size=batch_size)
```

```python
param_grid
```

#### perfomr 5-fold validation

```python
grid_search =  GridSearchCV(estimator=model,param_grid=param_grid,cv=5)
```

```python
results_1= grid_search.fit(X,y)
```

##### Print Bset cross validation accuracy Score

```python
print(results_1.best_score_*100)
```

##### Print best paramters for cross validation score

```python
print(results_1.best_params_)
```

##### print the results for all evaluated hyperparameter combinations

```python
accuracy_means = results_1.cv_results_['mean_test_score'] *100
print("Accuracy Means accross K-folds ",accuracy_means)
accuracy_std = results_1.cv_results_['std_test_score']
print("Accuracy_std",accuracy_std)
parameters = results_1.cv_results_['params']
print(" ",paramters,"\n")
```

```python

for p in range(len(paramters)):
    print("Accuracy %f (std %f) for params %r" %(accuracy_means[p], accuracy_std[p], parameters[p]))
```
